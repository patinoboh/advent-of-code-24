```python
import numpy as np
content = open("input.txt").read().split()
map_end = [i for i in range(len(content)) if all(x == "#" for x in content[i])][1]
m = np.asmatrix([list(line) for line in content[0:map_end+1]])
commands = "".join(content[map_end+1:])

mm = np.matrix([["." for _ in range(m.shape[0] * 2)]for _ in range(m.shape[1])])
mmm = np.matrix([[" " for _ in range(m.shape[0])]for _ in range(m.shape[1])])
for i in range(m.shape[0]):
    for j in range(m.shape[1]):
        if m[i,j] == ".":
            mm[i,2*j:2*j+2] = "."
        elif m[i,j] == "#":
            mm[i,2*j:2*j+2] = "#"
        elif m[i,j] == "O":
            mm[i,2*j] = "["
            mm[i,2*j+1] = "]"
        elif m[i,j] == "@":
            mm[i,2*j] = "@"
            mm[i,2*j+1] = "."
        mmm[i,j] = m[i,j]


i,j = np.argwhere(mm=="@")[0]

def print_m(m):
    for line in m.tolist():
        print("".join(line))

for command in commands:
    posi, posj = [(-1,0),(1,0),(0,-1),(0,1)]["^v<>".index(command)]
    if command in "<>":
        newi, newj = i+posi,j+posj
        while mm[newi,newj] in "[]":
            newi,newj = newi+posi,newj+posj
        if mm[newi, newj] == ".":
            if command == ">":
                # print("TOTO : ",mm[i, j:newj])
                # print("SEM ",mm[i,j+1:newj+1])
                mm[i,j+1:newj+1] = mm[i, j:newj]

            else:
                # print("toto ",mm[i,newj+1:j+1])
                # print("sem ",mm[i,newj:j])

                mm[i,newj:j] = mm[i,newj+1:j+1]
            mm[i,j] = "."
            i,j = i+posi,j+posj
    elif command in "^v":
        to_check = [(i,j)]
        to_move = []
        can_move = True

        while to_check and can_move:
            to_check_new = []
            for a,b in to_check:
                newi = a + posi
                if mm[newi,b] in "[":
                    to_check_new.append((newi,b))
                    to_check_new.append((newi,b+1))
                elif mm[newi,b] in "]":
                    to_check_new.append((newi,b))
                    to_check_new.append((newi,b-1))
                elif mm[newi,b] == "#":
                    can_move = False
                to_move.append((a,b))
            to_check = to_check_new
        if can_move:
            to_move = sorted(list(set(to_move)))
            to_move = to_move if command == "^" else to_move[::-1]
            # print("to move : ", to_move)
            for a,b in to_move:
                mm[a+posi,b] = mm[a,b]
                mm[a,b] = "."
            i += posi


result = 0
for i in range(mm.shape[0]):
    for j in range(mm.shape[1]):
        if mm[i,j] == '[':
            result += 100 * i + j

# print_m(mm) 
print(result)       
```
